```python
Yrs_ws = ['2000-01', '2001-02', '2002-03', '2003-04', '2004-05','2005-06', '2006-07', '2007-08', '2008-09',
          '2009-10', '2010-11','2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
Sect_ws = ['Sector']
```

#### Coversion of Investment which received in dollar from to indian rupee 
##### Exchange rate from year 2000 to 2017 required Reference Exchange rate history of the indian rupee from wikipedia
`Here in the reference there are two values of exchange rate that is annual average and end-year rate 
i have taken average exchange rate`

```python
# Exchange rate 2000-2017
ExchangeRt_Avg = [45.68,47.69,48.39,45.95,44.93,44.27,45.25,40.26,45.99,47.44,45.56,47.92,54.40,
                  60.50,61.14,65.47,67.07]
```

```python
FDI_spdUSD = FDI_spd.copy()
```

```python
# here  df is FDI_spd colm taken as year rts as exhange rate
def Convers_Usdinr(df,colm,rts):
    for ele in colm:
        df[ele] = df[ele] * rts[colm.index(ele)]/10
    return df    
```

```python
FDI_spdINR = Convers_Usdinr(FDI_spd,Yrs_ws,ExchangeRt_Avg)
```

```python
FDI_spdINR
```

```python
FDI_spdINR.style.set_caption("FDI Amount in INR Crores").format(precision=2)
```

### We will Unpivot the dataframe to get more broader picture of data  we will be using melt()

```python
melt = pd.melt(FDI_spdUSD, id_vars = Sect_ws, value_vars = Yrs_ws, var_name='Year', value_name='FDI(USD Million)',ignore_index=True)
melt
```

```python
melt=round(melt,2)
melt
```

```python
melt_01 = pd.melt(FDI_spdINR, id_vars = Sect_ws, value_vars = Yrs_ws, var_name='Year',value_name='FDI(INR Crores)',ignore_index=True)
melt_01
```

```python
# Round off two decimal point
melt_01=round(melt_01,2)
```

```python
melt_01
```

```python
# Merging the FDI(USD Million) column of melt Dataframe into melto1 Dataframe
Fdi_Merg= melt_01.merge(melt,how='left')
Fdi_Merg
```

```python
#Sorting the Sectors and Year columns
Fdi_Sorte = Fdi_Merg.sort_values(['Sector','Year'], ignore_index=True)
Fdi_Sorte
```

### We will Find Top 10 Sector received highest as well Bottom 10 Sector received lowest FDI `

# Top Sectors

```python
Sect_HighFdi = FDI_Sctrws.nlargest(10,['FDI(INR Crores)'])
Fdiin_Tot = round(melt_01['FDI(INR Crores)'].sum(),2)
Sum = Sect_HighFdi['FDI(INR Crores)'].sum()
Sect_HighFdi['In %age'] = round(Sect_HighFdi['FDI(INR Crores)']/Sum*100,2) 
Sect_HighFdi['%age to Total Inflows'] = round((Sect_HighFdi['FDI(INR Crores)']/Fdiin_Tot)*100,2)
```

```python
Sect_HighFdi
```

# Bottom Sectors

```python
Sect_LowFdi = FDI_Sctrws.nsmallest(10,['FDI(INR Crores)'])
Sum = Sect_LowFdi['FDI(INR Crores)'].sum()
Sect_LowFdi['In %age'] = round(Sect_LowFdi['FDI(INR Crores)']/Sum*100,2)
Sect_LowFdi['%age to Total Inflows'] = round((Sect_LowFdi['FDI(INR Crores)']/Fdiin_Tot)*100,3)
```

```python
Sect_LowFdi
```
